```python
import numpy as np 
import pandas as pd
```

```python
data=pd.read_csv("https://raw.githubusercontent.com/reddyprasade/Machine-Learning-Problems-DataSets/master/Classification/mushrooms.csv")
```

```python
data.head()
```

```python
data.tail()
```

```python
#Set all columns as categorical. One hot encode them
for col in data.columns :
    data[col] = pd.get_dummies(data[col])
```

```python
#One hot encoded data
data.head()
```

```python
#X and Y
X = data.drop(['class'],axis=1)
y = data['class']
```

```python
import seaborn as sns
sns.countplot(x='class',data=data)
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=0.3, random_state=0)
```

```python
from sklearn.ensemble import RandomForestClassifier
rclf = RandomForestClassifier()
```

```python
rclf.fit(X_train,y_train)
```

```python
from sklearn.metrics import accuracy_score,recall_score,confusion_matrix
y_pred = rclf.predict(X_test)
print(len(X_test))
print(accuracy_score(y_test,y_pred))
print(recall_score(y_test,y_pred))
print(confusion_matrix(y_test,y_pred))
```
